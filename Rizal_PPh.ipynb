```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from io import StringIO

# Loading data from CSV strings (simulating file reading in Colab)
aset_tetap = """aset_id,kategori,nilai_perolehan,umur_ekonomis,metode
A001,Mesin Produksi,100000000,10,garis_lurus
A002,Kendaraan,80000000,8,saldo_menurun
A003,Peralatan Kantor,20000000,5,garis_lurus"""

kebijakan_fiskal = """tahun,tax_rate,tax_holiday_awal,tax_holiday_akhir
2023,0.22,2023,2027"""

transaksi_keuangan = """tahun,pendapatan,beban_operasional,penyusutan,skenario
2023-01,20000000,27000000,2500000,normal
2023-02,25000000,28800000,2500000,normal
2023-03,30000000,30800000,2500000,normal
2023-04,35000000,32000000,2500000,normal
2023-05,40000000,33500000,2500000,normal
2023-06,45000000,35000000,2500000,normal
2023-07,50000000,37000000,2500000,normal
2023-08,55000000,38500000,2500000,normal
2023-09,60000000,40000000,2500000,normal
2023-10,65000000,41500000,2500000,normal
2023-11,70000000,43000000,2500000,normal
2023-12,75000000,44500000,2500000,normal
2023-01,20000000,27000000,2500000,holiday
2023-02,25000000,28800000,2500000,holiday
2023-03,30000000,30800000,2500000,holiday
2023-04,35000000,32000000,2500000,holiday
2023-05,40000000,33500000,2500000,holiday
2023-06,45000000,35000000,2500000,holiday
2023-07,50000000,37000000,2500000,holiday
2023-08,55000000,38500000,2500000,holiday
2023-09,60000000,40000000,2500000,holiday
2023-10,65000000,41500000,2500000,holiday
2023-11,70000000,43000000,2500000,holiday
2023-12,75000000,44500000,2500000,holiday"""

# Converting CSV strings to DataFrames
df_aset = pd.read_csv(StringIO(aset_tetap))
df_fiskal = pd.read_csv(StringIO(kebijakan_fiskal))
df_transaksi = pd.read_csv(StringIO(transaksi_keuangan))

# Calculating depreciation for straight-line and double-declining balance methods
def calculate_depreciation(asset_value, useful_life, method, year):
    if method == 'garis_lurus':
        return asset_value / useful_life
    elif method == 'saldo_menurun':
        return (2 * asset_value / useful_life) * (1 - 2 / useful_life) ** (year - 2023)
    return 0

# Creating depreciation scenario DataFrame
df_depreciation = df_transaksi[df_transaksi['skenario'] == 'normal'].copy()
df_depreciation['penyusutan_garis_lurus'] = 0
df_depreciation['penyusutan_saldo_menurun'] = 0

# Calculating total depreciation per asset for each method
for _, row in df_aset.iterrows():
    if row['metode'] == 'garis_lurus':
        df_depreciation['penyusutan_garis_lurus'] += row['nilai_perolehan'] / row['umur_ekonomis'] / 12
    else:
        df_depreciation['penyusutan_saldo_menurun'] += (2 * row['nilai_perolehan'] / row['umur_ekonomis'] / 12)

# Calculating taxable income and PPh for each scenario
df_transaksi['laba_sebelum_pajak'] = df_transaksi['pendapatan'] - df_transaksi['beban_operasional'] - df_transaksi['penyusutan']
df_transaksi['pph'] = df_transaksi.apply(
    lambda row: 0 if row['skenario'] == 'holiday' else row['laba_sebelum_pajak'] * df_fiskal['tax_rate'].iloc[0]
    if row['laba_sebelum_pajak'] > 0 else 0, axis=1)

df_depreciation['laba_sebelum_pajak_garis_lurus'] = df_transaksi[df_transaksi['skenario'] == 'normal']['pendapatan'] - \
                                                    df_transaksi[df_transaksi['skenario'] == 'normal']['beban_operasional'] - \
                                                    df_depreciation['penyusutan_garis_lurus']
df_depreciation['laba_sebelum_pajak_saldo_menurun'] = df_transaksi[df_transaksi['skenario'] == 'normal']['pendapatan'] - \
                                                     df_transaksi[df_transaksi['skenario'] == 'normal']['beban_operasional'] - \
                                                     df_depreciation['penyusutan_saldo_menurun']
df_depreciation['pph_garis_lurus'] = df_depreciation['laba_sebelum_pajak_garis_lurus'].apply(
    lambda x: x * df_fiskal['tax_rate'].iloc[0] if x > 0 else 0)
df_depreciation['pph_saldo_menurun'] = df_depreciation['laba_sebelum_pajak_saldo_menurun'].apply(
    lambda x: x * df_fiskal['tax_rate'].iloc[0] if x > 0 else 0)

# Aggregating data for visualization
monthly_summary = df_transaksi.groupby(['tahun', 'skenario'])['pph'].sum().unstack().reset_index()
depreciation_summary = df_depreciation.groupby('tahun')[['pph_garis_lurus', 'pph_saldo_menurun']].sum().reset_index()

# Visualizing the results
plt.figure(figsize=(12, 8))

# Plot 1: PPh for Normal vs Tax Holiday scenarios
plt.subplot(2, 1, 1)
plt.plot(monthly_summary['tahun'], monthly_summary['normal'] / 1_000_000, label='PPh Normal (Juta)', color='blue')
plt.plot(monthly_summary['tahun'], monthly_summary['holiday'] / 1_000_000, label='PPh Tax Holiday (Juta)', color='green')
plt.title('Perbandingan PPh: Normal vs Tax Holiday')
plt.xlabel('Bulan')
plt.ylabel('PPh (Juta Rupiah)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# Plot 2: PPh for different depreciation methods
plt.subplot(2, 1, 2)
plt.plot(depreciation_summary['tahun'], depreciation_summary['pph_garis_lurus'] / 1_000_000, label='PPh Garis Lurus (Juta)', color='blue')
plt.plot(depreciation_summary['tahun'], depreciation_summary['pph_saldo_menurun'] / 1_000_000, label='PPh Saldo Menurun (Juta)', color='orange')
plt.title('Perbandingan PPh: Metode Depresiasi')
plt.xlabel('Bulan')
plt.ylabel('PPh (Juta Rupiah)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Generating analytical narrative
narrative = """
# Analisis Keuangan dan Pajak Penghasilan (PPh)

## Ringkasan
Analisis ini mengevaluasi Pajak Penghasilan (PPh) berdasarkan tiga skenario: Normal, Tax Holiday, dan Perbandingan Metode Depresiasi (Garis Lurus vs Saldo Menurun) menggunakan data aset tetap, kebijakan fiskal, dan transaksi keuangan tahun 2023.

## Visualisasi
1. **Normal vs Tax Holiday**:
   - Skenario Normal menghasilkan PPh yang signifikan karena dikenakan tarif pajak 22%.
   - Skenario Tax Holiday menghasilkan PPh nol karena berada dalam periode tax holiday (2023-2027).
   - Grafik menunjukkan PPh bulanan pada skenario Normal meningkat seiring kenaikan pendapatan, sementara Tax Holiday tetap nol.

2. **Perbandingan Metode Depresiasi**:
   - Metode Garis Lurus menghasilkan penyusutan yang konsisten setiap bulan.
   - Metode Saldo Menurun menghasilkan penyusutan lebih tinggi di awal, sehingga mengurangi laba kena pajak dan PPh lebih besar di bulan-bulan awal.
   - Grafik menunjukkan PPh dengan metode Saldo Menurun lebih rendah dibandingkan Garis Lurus karena penyusutan yang lebih agresif.

## Fakta Menarik
Metode Saldo Menurun memberikan keuntungan pajak yang lebih besar di awal periode karena penyusutan yang lebih tinggi, yang dapat meningkatkan arus kas perusahaan di tahun-tahun awal.

## Evaluasi
- **Skenario Normal**: PPh yang tinggi mengurangi laba bersih, terutama pada bulan dengan pendapatan tinggi.
- **Skenario Tax Holiday**: Menguntungkan karena tidak ada beban PPh, memungkinkan perusahaan mengalokasikan lebih banyak dana untuk investasi atau operasional.
- **Metode Depresiasi**: Saldo Menurun lebih menguntungkan untuk mengurangi PPh di awal, tetapi Garis Lurus memberikan stabilitas dalam perencanaan keuangan jangka panjang.

## Interpretasi Hasil
- Skenario Tax Holiday memberikan manfaat fiskal yang signifikan, terutama untuk perusahaan dengan investasi besar dalam aset tetap.
- Pilihan metode depresiasi memengaruhi arus kas dan kewajiban pajak tahunan. Saldo Menurun cocok untuk perusahaan yang ingin memaksimalkan penghematan pajak di awal, sedangkan Garis Lurus lebih cocok untuk perencanaan yang stabil.

## Keterkaitan Keputusan Fiskal dan Dampaknya
- **Kebijakan Fiskal**: Tax holiday mendorong investasi dengan mengurangi beban pajak, meningkatkan likuiditas perusahaan.
- **Dampak pada Bisnis**: Penghematan dari tax holiday dapat digunakan untuk ekspansi atau inovasi. Pilihan metode depresiasi memengaruhi strategi keuangan jangka pendek dan panjang, dengan Saldo Menurun memberikan manfaat arus kas awal yang lebih besar.

## Kesimpulan
Skenario Tax Holiday adalah pilihan terbaik untuk meminimalkan PPh dan meningkatkan arus kas. Namun, pemilihan metode depresiasi harus disesuaikan dengan tujuan keuangan perusahaan, dengan Saldo Menurun lebih cocok untuk penghematan pajak awal dan Garis Lurus untuk stabilitas jangka panjang.
"""

# Displaying narrative in Colab
from IPython.display import Markdown, display
display(Markdown(narrative))

# Saving narrative to a markdown file (optional for download)
with open('financial_narrative.md', 'w') as f:
    f.write(narrative)

# Printing confirmation
print("Analisis selesai. Visualisasi ditampilkan di atas, dan narasi tersimpan sebagai 'financial_narrative.md'.")
import matplotlib.pyplot as plt
import numpy as np
from io import StringIO

# Loading data from CSV strings (simulating file reading in Colab)
aset_tetap = """aset_id,kategori,nilai_perolehan,umur_ekonomis,metode
A001,Mesin Produksi,100000000,10,garis_lurus
A002,Kendaraan,80000000,8,saldo_menurun
A003,Peralatan Kantor,20000000,5,garis_lurus"""

kebijakan_fiskal = """tahun,tax_rate,tax_holiday_awal,tax_holiday_akhir
2023,0.22,2023,2027"""

transaksi_keuangan = """tahun,pendapatan,beban_operasional,penyusutan,skenario
2023-01,20000000,27000000,2500000,normal
2023-02,25000000,28800000,2500000,normal
2023-03,30000000,30800000,2500000,normal
2023-04,35000000,32000000,2500000,normal
2023-05,40000000,33500000,2500000,normal
2023-06,45000000,35000000,2500000,normal
2023-07,50000000,37000000,2500000,normal
2023-08,55000000,38500000,2500000,normal
2023-09,60000000,40000000,2500000,normal
2023-10,65000000,41500000,2500000,normal
2023-11,70000000,43000000,2500000,normal
2023-12,75000000,44500000,2500000,normal
2023-01,20000000,27000000,2500000,holiday
2023-02,25000000,28800000,2500000,holiday
2023-03,30000000,30800000,2500000,holiday
2023-04,35000000,32000000,2500000,holiday
2023-05,40000000,33500000,2500000,holiday
2023-06,45000000,35000000,2500000,holiday
2023-07,50000000,37000000,2500000,holiday
2023-08,55000000,38500000,2500000,holiday
2023-09,60000000,40000000,2500000,holiday
2023-10,65000000,41500000,2500000,holiday
2023-11,70000000,43000000,2500000,holiday
2023-12,75000000,44500000,2500000,holiday"""

# Converting CSV strings to DataFrames
df_aset = pd.read_csv(StringIO(aset_tetap))
df_fiskal = pd.read_csv(StringIO(kebijakan_fiskal))
df_transaksi = pd.read_csv(StringIO(transaksi_keuangan))

# Calculating depreciation for straight-line and double-declining balance methods
def calculate_depreciation(asset_value, useful_life, method, year):
    if method == 'garis_lurus':
        return asset_value / useful_life
    elif method == 'saldo_menurun':
        return (2 * asset_value / useful_life) * (1 - 2 / useful_life) ** (year - 2023)
    return 0

# Creating depreciation scenario DataFrame
df_depreciation = df_transaksi[df_transaksi['skenario'] == 'normal'].copy()
df_depreciation['penyusutan_garis_lurus'] = 0
df_depreciation['penyusutan_saldo_menurun'] = 0

# Calculating total depreciation per asset for each method
for _, row in df_aset.iterrows():
    if row['metode'] == 'garis_lurus':
        df_depreciation['penyusutan_garis_lurus'] += row['nilai_perolehan'] / row['umur_ekonomis'] / 12
    else:
        df_depreciation['penyusutan_saldo_menurun'] += (2 * row['nilai_perolehan'] / row['umur_ekonomis'] / 12)

# Calculating taxable income and PPh for each scenario
df_transaksi['laba_sebelum_pajak'] = df_transaksi['pendapatan'] - df_transaksi['beban_operasional'] - df_transaksi['penyusutan']
df_transaksi['pph'] = df_transaksi.apply(
    lambda row: 0 if row['skenario'] == 'holiday' else row['laba_sebelum_pajak'] * df_fiskal['tax_rate'].iloc[0]
    if row['laba_sebelum_pajak'] > 0 else 0, axis=1)

df_depreciation['laba_sebelum_pajak_garis_lurus'] = df_transaksi[df_transaksi['skenario'] == 'normal']['pendapatan'] - \
                                                    df_transaksi[df_transaksi['skenario'] == 'normal']['beban_operasional'] - \
                                                    df_depreciation['penyusutan_garis_lurus']
df_depreciation['laba_sebelum_pajak_saldo_menurun'] = df_transaksi[df_transaksi['skenario'] == 'normal']['pendapatan'] - \
                                                     df_transaksi[df_transaksi['skenario'] == 'normal']['beban_operasional'] - \
                                                     df_depreciation['penyusutan_saldo_menurun']
df_depreciation['pph_garis_lurus'] = df_depreciation['laba_sebelum_pajak_garis_lurus'].apply(
    lambda x: x * df_fiskal['tax_rate'].iloc[0] if x > 0 else 0)
df_depreciation['pph_saldo_menurun'] = df_depreciation['laba_sebelum_pajak_saldo_menurun'].apply(
    lambda x: x * df_fiskal['tax_rate'].iloc[0] if x > 0 else 0)

# Aggregating data for visualization
monthly_summary = df_transaksi.groupby(['tahun', 'skenario'])['pph'].sum().unstack().reset_index()
depreciation_summary = df_depreciation.groupby('tahun')[['pph_garis_lurus', 'pph_saldo_menurun']].sum().reset_index()

# Visualizing the results
plt.figure(figsize=(12, 8))

# Plot 1: PPh for Normal vs Tax Holiday scenarios
plt.subplot(2, 1, 1)
plt.plot(monthly_summary['tahun'], monthly_summary['normal'] / 1_000_000, label='PPh Normal (Juta)', color='blue')
plt.plot(monthly_summary['tahun'], monthly_summary['holiday'] / 1_000_000, label='PPh Tax Holiday (Juta)', color='green')
plt.title('Perbandingan PPh: Normal vs Tax Holiday')
plt.xlabel('Bulan')
plt.ylabel('PPh (Juta Rupiah)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# Plot 2: PPh for different depreciation methods
plt.subplot(2, 1, 2)
plt.plot(depreciation_summary['tahun'], depreciation_summary['pph_garis_lurus'] / 1_000_000, label='PPh Garis Lurus (Juta)', color='blue')
plt.plot(depreciation_summary['tahun'], depreciation_summary['pph_saldo_menurun'] / 1_000_000, label='PPh Saldo Menurun (Juta)', color='orange')
plt.title('Perbandingan PPh: Metode Depresiasi')
plt.xlabel('Bulan')
plt.ylabel('PPh (Juta Rupiah)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Generating analytical narrative
narrative = """
# Analisis Keuangan dan Pajak Penghasilan (PPh)

## Ringkasan
Analisis ini mengevaluasi Pajak Penghasilan (PPh) berdasarkan tiga skenario: Normal, Tax Holiday, dan Perbandingan Metode Depresiasi (Garis Lurus vs Saldo Menurun) menggunakan data aset tetap, kebijakan fiskal, dan transaksi keuangan tahun 2023.

## Visualisasi
1. **Normal vs Tax Holiday**:
   - Skenario Normal menghasilkan PPh yang signifikan karena dikenakan tarif pajak 22%.
   - Skenario Tax Holiday menghasilkan PPh nol karena berada dalam periode tax holiday (2023-2027).
   - Grafik menunjukkan PPh bulanan pada skenario Normal meningkat seiring kenaikan pendapatan, sementara Tax Holiday tetap nol.

2. **Perbandingan Metode Depresiasi**:
   - Metode Garis Lurus menghasilkan penyusutan yang konsisten setiap bulan.
   - Metode Saldo Menurun menghasilkan penyusutan lebih tinggi di awal, sehingga mengurangi laba kena pajak dan PPh lebih besar di bulan-bulan awal.
   - Grafik menunjukkan PPh dengan metode Saldo Menurun lebih rendah dibandingkan Garis Lurus karena penyusutan yang lebih agresif.

## Fakta Menarik
Metode Saldo Menurun memberikan keuntungan pajak yang lebih besar di awal periode karena penyusutan yang lebih tinggi, yang dapat meningkatkan arus kas perusahaan di tahun-tahun awal.

## Evaluasi
- **Skenario Normal**: PPh yang tinggi mengurangi laba bersih, terutama pada bulan dengan pendapatan tinggi.
- **Skenario Tax Holiday**: Menguntungkan karena tidak ada beban PPh, memungkinkan perusahaan mengalokasikan lebih banyak dana untuk investasi atau operasional.
- **Metode Depresiasi**: Saldo Menurun lebih menguntungkan untuk mengurangi PPh di awal, tetapi Garis Lurus memberikan stabilitas dalam perencanaan keuangan jangka panjang.

## Interpretasi Hasil
- Skenario Tax Holiday memberikan manfaat fiskal yang signifikan, terutama untuk perusahaan dengan investasi besar dalam aset tetap.
- Pilihan metode depresiasi memengaruhi arus kas dan kewajiban pajak tahunan. Saldo Menurun cocok untuk perusahaan yang ingin memaksimalkan penghematan pajak di awal, sedangkan Garis Lurus lebih cocok untuk perencanaan yang stabil.

## Keterkaitan Keputusan Fiskal dan Dampaknya
- **Kebijakan Fiskal**: Tax holiday mendorong investasi dengan mengurangi beban pajak, meningkatkan likuiditas perusahaan.
- **Dampak pada Bisnis**: Penghematan dari tax holiday dapat digunakan untuk ekspansi atau inovasi. Pilihan metode depresiasi memengaruhi strategi keuangan jangka pendek dan panjang, dengan Saldo Menurun memberikan manfaat arus kas awal yang lebih besar.

## Kesimpulan
Skenario Tax Holiday adalah pilihan terbaik untuk meminimalkan PPh dan meningkatkan arus kas. Namun, pemilihan metode depresiasi harus disesuaikan dengan tujuan keuangan perusahaan, dengan Saldo Menurun lebih cocok untuk penghematan pajak awal dan Garis Lurus untuk stabilitas jangka panjang.
"""

# Displaying narrative in Colab
from IPython.display import Markdown
display(Markdown(narrative))

# Saving narrative to a markdown file (optional for download)
with open('financial_narrative.md', 'w') as f:
    f.write(narrative)

# Printing confirmation
print("Analisis selesai. Visualisasi ditampilkan di atas, dan narasi tersimpan sebagai 'financial_narrative.md'.")
```
